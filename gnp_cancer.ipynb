```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
#fig = px.histogram(data, x="diagnosis", title="Sum of diagnosis")
#fig.update_layout(bargap=0.1)
#fig.show()
ax = sns.countplot(x='diagnosis', data=data)
plt.figure(figsize=(10,10))
plt.show()
```

```python
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]
plt.title("Malignant vs Benign Tumor")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.scatter(M.radius_mean, M.texture_mean, color = "red", label = "Malignant", alpha = 0.3)
plt.scatter(B.radius_mean, B.texture_mean, color = "lime", label = "Benign", alpha = 0.3)
plt.legend()
plt.show()
```

# Tahap Klasifikasi

```python
#prep1
#normalisasi
diag_newlist = []
 
for n in data.diagnosis:
    if n == "M":
        diag_newlist.append(1)
    else:
        diag_newlist.append(0)
#diag_newlist

#print(diag_newlist)
#col_diag_list = data['diagnosis'].tolist()
#col_one_arr = df['one'].to_numpy()
#print(col_diag_list)
import numpy as np
data.diagnosis = diag_newlist
#Prep_2
x = data.drop(["diagnosis"], axis = 1)
y = data.diagnosis.values


#TrainSplit (80:20)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

#GausianNB
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)

data_model = GaussianNB() #output
print("#=============================================================#")
#akurasi
print("Naive Bayes score: ",nb.score(x_test, y_test))

NB_train = data_model.fit(x_train, y_train)

y_pred = NB_train.predict(x_test)
np.array(y_pred) #hasil prediksi 

np.array(y_test) #y_test separasi data
print("#=============================================================#")
#confusion matrix berdasar hasil prediksi
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, y_pred) 
print("#=============================================================#")
#evaluasi performansi dari confusion
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print("#=============================================================#")
print(confusion_matrix(y_test, y_pred))
print("#=============================================================#")
#penampilan semua probability
NB_train.predict_proba(x_test)
```

```python
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List harus alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## display confusion matrix
plt.show()
```
